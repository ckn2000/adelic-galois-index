```python
from lmf import db
```

```python
ec_galrep = db["ec_galrep"]
```

```python
def get_galrep(label):
    # Each record has a prime value (0 for adelic),
    # others correspond to mod-p Galois images.
    return list(ec_galrep.search({"lmfdb_label": label}))
```

```python
labels = []
coeffs = []
levels = []
```

```python
with open("2,2.txt") as f:
    for line in f:
        line = line.strip()
        if not line or line.startswith("#"):
            continue
        parts = line.split("\t")
        label = parts[0].strip('"')
        coefficient = parts[-1].strip('"')
        level = parts[1].strip('"')
        labels.append(label)
        coeffs.append(coefficient)
        levels.append(level)
```

```python
adelic_list = []
```

```python
for i, label in enumerate(labels):
    galrep = get_galrep(label)[0]
    adelic_gens = galrep["adelic_gens"]
    adelic_list.append((label,coeffs[i],levels[i],adelic_gens))
```

```python
with open("index2,2.txt", "w") as f:
    for entry in adelic_list:
        label, coeff, level, gens = entry
        gens_str = str(gens).replace(" ", "")
        f.write(f"{label};{coeff};{gens_str};{level}\n")
```
